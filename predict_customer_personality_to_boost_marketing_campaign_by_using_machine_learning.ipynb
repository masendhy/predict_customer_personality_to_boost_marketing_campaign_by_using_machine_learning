### **Customer Personality Analysis**
Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

[Download Resources here](https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis)



### Dataset Column Detail

`NumWebPurchases` | Number of purchases made through the company website
`NumWebVisitsMonth` | Number of visits to the company's website in the past month
`Recency` | Number of days since last purchase

```python
df_featex = df.copy()
```

```python
#convert Dt_Customer to datetime
from datetime import datetime
df_featex['Dt_Customer'] = pd.to_datetime(df_featex['Dt_Customer'],format='%d-%m-%Y') 
```

```python
# create years joined column
df_featex['YearsJoined'] = datetime.now().year - df_featex['Dt_Customer'].dt.year
```

```python
#created total amount a customer has spent
df_featex['TotalSpent'] = df_featex['MntCoke'] + df_featex['MntFruits'] + df_featex['MntMeatProducts'] + df_featex['MntFishProducts'] + df_featex['MntSweetProducts'] + df_featex['MntGoldProds']
```

```python
# check average customer's income spent
df_featex.income_to_spent = round(df_featex['Income']/df_featex['TotalSpent'],2)
 
```

```python
#create Age column
df_featex['Age'] = 2023 - df_featex['Year_Birth']
```

```python
df_featex['Total_AcceptedCmp'] = df_featex['AcceptedCmp3'] + df_featex['AcceptedCmp4'] + df_featex['AcceptedCmp5'] + df_featex['AcceptedCmp1'] + df_featex['AcceptedCmp2']
```

```python
df_featex['Total_Purchase'] = df_featex['NumDealsPurchases'] + df_featex['NumStorePurchases'] + df_featex['NumWebPurchases'] + df_featex['NumCatalogPurchases']  
```

```python
# grouping Age
age_list = []
for i in df_featex['Age']:
    if i >= 0 and i <= 9:
        group = '1'
    elif i >= 11 and i <= 19:
        group = '2'
    elif i >= 20 and i <= 35:
        group = '3'
    elif i >= 36 and i <= 49:
        group = '4'
    else:
        group = '5'
    age_list.append(group)
df_featex['Age_Group'] = age_list
```

```python
# create amount  of children column
df_featex['NumChildren'] = df_featex['Kidhome'] + df_featex['Teenhome']
```

```python
df_featex['Marital_Status'].value_counts()
```

```python
# grouping marital status 
married = ['Menikah','Bertunangan']
unmarried = ['Lajang','Cerai','Janda','Duda']

marital_list = []
for i in df_featex['Marital_Status']:   
    if i in unmarried:
        group = '0'
    else:
        group = '1'
    marital_list.append(group)
df_featex['Marital_List'] = marital_list
```

### **Data Modeling**

### *Clustering*
In this case I’ll try using K-means and agglomerative clustering

```python
# lets do clustering using K-Means
from sklearn.cluster import KMeans

wcss=[]
for i in range(1,9):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df_model)
    wcss.append(kmeans.inertia_)
```

```python
# Let’s see how many clustering is the best by seeing it with the elbow method

plt.figure(figsize=(10,8))

plt.plot(range(1,9),wcss,marker = 'o',
         markersize = 10,linewidth=4,markerfacecolor = 'red')

for i, label in enumerate(wcss):
    plt.text(range(1,9)[i] + 0.2,wcss[i] + 7,label ,ha='center',va='center')


plt.xlabel('Number of Clustering')
plt.ylabel('Inertia')
plt.title('K-means Clustering')
plt.show()
```

Ok, 4 is the best clustering based on elbow method, because wcss score did'nt change significanly

```python
# Let’s see the optimum number of clustering based on the silhouette score
from sklearn.metrics import silhouette_score

silhouette_score_arr = []
for i in range(2,11):
    kmean = KMeans(n_clusters = i).fit(df_model)
    predc = kmean.predict(df_model)
    score = silhouette_score(df_model,predc)
    silhouette_score_arr.append(np.round(score,2))
    print("On k =",i,"silhouette_score = ",score)
```

```python
# show visualization of silhouette score

x = range(2,11)

plt.figure(figsize=(10,8))
plt.plot(range(2,11),silhouette_score_arr,marker = 'o',
         markersize = 10,linewidth=4,markerfacecolor = 'red')

sns.scatterplot(x=x,y=silhouette_score_arr)

#give text label for shilouette score
for i, label in enumerate(silhouette_score_arr):
    plt.text(x[i], label, str(label))
```

*when n_cluster = 4 , get silhoutte score = 0.23*

### **Train Model**

```python
# let's train the model
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4,init='k-means++',random_state=42)
kmeans.fit(df_model)


df_model['cluster'] = kmeans.labels_
df_cluster['cluster'] = kmeans.labels_
```

```python
df_cluster.sample(9)
```

### **Principal Component Analysis ( PCA)**
A data analysis technique used to reduce the dimensionality of data by preserving as much important information as possible.

```python
from sklearn.decomposition import PCA


df_pca =  df_model.drop(['cluster'],axis=1)

pca = PCA(n_components=2)
pca.fit(df_pca)
df_pctransform = pca.transform(df_pca)
```

```python
pca_transformed = pd.DataFrame(data=df_pctransform,columns=['PC1','PC2'])
pca_transformed
```

```python
pca_transformed['cluster'] = df_cluster['cluster']
pca_transformed
```

```python
#lets show the visualization of PCA using scatter plot
fig,ax = plt.subplots(figsize=(15,10))
sns.scatterplot(x='PC1',y='PC2',hue='cluster',data=pca_transformed,palette="pastel" ,s=100)
plt.title('PCA 2D Visualisation',fontsize=20,weight='bold')
```

```python
# create 3D visualisation of PCA

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(df_cluster['Recency'][df_cluster.cluster==2 ], df_cluster['TotalSpent'][df_cluster.cluster==2 ], df_cluster['Total_AcceptedCmp'][df_cluster.cluster==2 ],c='blue', s=100,edgecolors='black',  label='High-Spender Customer')
ax.scatter(df_cluster['Recency'][df_cluster.cluster==1 ], df_cluster['TotalSpent'][df_cluster.cluster==1 ], df_cluster['Total_AcceptedCmp'][df_cluster.cluster==1 ],c='red', s=100,edgecolors='black',  label='Mid-Spender Customer')
ax.scatter(df_cluster['Recency'][df_cluster.cluster==3 ], df_cluster['TotalSpent'][df_cluster.cluster==3 ], df_cluster['Total_AcceptedCmp'][df_cluster.cluster==3 ],c='green', s=100,edgecolors='black',  label='Low-Spender Customer')
ax.scatter(df_cluster['Recency'][df_cluster.cluster==0 ], df_cluster['TotalSpent'][df_cluster.cluster==0 ], df_cluster['Total_AcceptedCmp'][df_cluster.cluster==0 ],c='yellow', s=100,edgecolors='black',  label='Risk_of_Churn Customer')

plt.title('PCA 3D Visualisation',fontsize=20,weight='bold')
plt.xlabel('Recency')
plt.ylabel('Total Purchases')
ax.set_zlabel('Total Spent')
plt.legend(title='Cluster:')
plt.show()
```
